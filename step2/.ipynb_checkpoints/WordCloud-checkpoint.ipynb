# Step2 统计词频与词云图生成

    统计不同类别的新闻的高频词，并以词云的形式展示。
    顺序为：
        导入所需的库 --> 导入文件 --> 切分关键词 --> 统计计数 --> 导入词云 -> 结果导出

# 1. 导入所需的库

```python
# -*- coding: utf-8 -*-
import jieba
import numpy as np
import codecs
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud
```

# 2. 从文件中导入数据

    这里以DataFrame的格式导入文件。
    注：其实是因为没有按照csv格式存储又不想重新爬一遍。

```python
import json

def readfromJson(filename):
    dir = "data/"
    filename = dir + filename
    file = codecs.open(filename, 'r')
    
    data = {}
    data['title'] = []
    data['publisher'] = []
    data['type'] = []
    data['time'] = []
    data['content'] = []
    
    for line in file.readlines():
        news_unit = json.loads(line)
        data['title'].append(news_unit['title'])
        data['publisher'].append(news_unit['publisher'])
        data['type'].append(news_unit['type'])
        data['time'].append(news_unit['time'])
        data['content'].append(news_unit['content'])

    data = pd.DataFrame(data)
    return data

GuojiData = readfromJson("Guoji.json")
GuoneiData = readfromJson("Guonei.json")
ShehuiData = readfromJson("Shehui.json")
frames = [GuojiData, GuoneiData, ShehuiData]
data = pd.concat(frames)
```

# 3. 切分关键词

    切分关键词，然后将关键词读入一个list中。

```python
data["segment"] = data["content"].apply(lambda x : list(jieba.cut(x)))
tmp = list(data["segment"])
segments = []
for content in tmp:
    segments.extend(content)
data = data.drop(['segment'], axis=1)
```

# 4.过滤停用词与统计计数

    过滤停用词。
    停用词表足够全面，包含了常用的汉语停用词。
    当然还要去掉空格回车符等字符。

```python
words_df = pd.DataFrame({'segment' : segments})
words_df = words_df.dropna()

stopwords = pd.read_csv("stopwords.txt", index_col=False, quoting=3, sep="\t", names=['stopword'], encoding = "utf-8")
words_df = words_df[~words_df.segment.isin(stopwords.stopword)]
words_df = words_df[~words_df.segment.isin(['\n', '\r', ' '])]
words_df.head(10)
```

    按计数统计，然后排序，得到高频词表。

```python
words_stat = words_df.groupby(by = ['segment'])['segment'].agg({"计数" : np.size})
words_stat = words_stat.reset_index().sort(columns = "计数",ascending = False)
words_stat.head(10)
```

# 5.做词云图

    直接掉包导入词云图，词频大的字体大小显著。

```python
wordcloud = WordCloud(font_path = "simhei.ttf",background_color = "black")
#以计数值为权重，取前1000项生成词云图
wordcloud = wordcloud.fit_words(words_stat.head(1000).itertuples(index=False))
plt.imshow(wordcloud)
```

    词的大小差距显著。考虑对计数做平滑处理。

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
word_stat_std = sc.fit_transform(words_stat['计数'])
words_stat['count_std'] = word_stat_std
words_stat_std = words_stat[['segment', 'count_std']]
words_stat_std.head(10)
```

    词的大小没有发生变化。

```python
wordcloud = WordCloud(font_path = "simhei.ttf",background_color = "black")
#以计数值为权重，取前1000项生成词云图
wordcloud = wordcloud.fit_words(words_stat_std.head(1000).itertuples(index=False))
plt.imshow(wordcloud)
```

# 6.结果保存

    结果保存为csv。

```python
words_stat.set_index("segment", inplace = True)
words_stat.to_csv('result.csv', encoding = "utf-8")
```
